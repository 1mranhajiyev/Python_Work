```python
import pandas as pd 
import numpy as np
df=pd.read_csv('spot_2.csv',parse_dates=['block_timestamp'])
df=df.sort_values(by=['block_timestamp','transaction_index','log_index'])
df=df.reset_index()
df = df.drop(columns='index')
df
```

```python
df['token0']=df['token0_in'].fillna(0)+df['token0_out'].fillna(0)
df['token1']=df['token1_in'].fillna(0)+df['token1_out'].fillna(0)
df['Spotprice']=np.where(
        (df['event_name'] == 'Swap'), 
         (df['token0']/df['token1']), 
         np.where(
    (df['event_name'] == 'Increase liquidity') & (df['event_name'].shift(1) == 'Initialize'),
    (df['token0'] / df['token1']),np.nan))
df['Spotprice'] = df['Spotprice'].ffill()
df['Volume'] = np.where(
    (df['event_name'] == 'Swap') & (df['token0_in'].isna()),
    df['token1_in']*df['Spotprice'],
    np.where(
        (df['event_name'] == 'Swap') & (df['token1_in'].isna()), 
        df['token0_in'],np.nan)
        )
df['fee']=df['Volume']*0.003
df1 = df[df['event_name'] != 'Burn (Remove)']
df['Tvl']=np.cumsum((df1.fillna(0)['token0_in'] - df1.fillna(0)['token0_out']) + (df1.fillna(0)['token1_in'] - df1.fillna(0)['token1_out']) *df1.Spotprice)
# df1=df.resample('D',on='block_timestamp')[['fee','Volume','Tvl']].mean().reset_index()
df['Apy']=(((1+df['fee']/df['Tvl'])**365)-1)*100
df['Apr']=(df['fee']/df['Tvl'])*365*100
df
```

```python
df['token0']=df['token0_in'].fillna(0)+df['token0_out'].fillna(0) 
df['token1']=df['token1_in'].fillna(0)+df['token1_out'].fillna(0)
swap_mask = (df['event_name'] == 'Swap')
liquidity_mask = (df['event_name'] == 'Increase liquidity') & (df['event_name'].shift(1) == 'Initialize')

df['Spotprice'] = np.where(swap_mask, df['token0'] / df['token1'], 
                           np.where(liquidity_mask, df['token0'] / df['token1'], np.nan))
df['Spotprice'] = df['Spotprice'].ffill()

mask_1 = (df['event_name'] == 'Swap') & (df['token0_in'].isna())
mask_2 = (df['event_name'] == 'Swap') & (df['token1_in'].isna())
df['Volume'] = np.where(mask_1, df['token1_in'] * df['Spotprice'], 
                        np.where(mask_2, df['token0_in'], np.nan))

df['fee']=df['Volume']*0.003

df1 = df[df['event_name'] != 'Burn (Remove)']
df['Tvl']=np.cumsum((df1.fillna(0)['token0_in'] - df1.fillna(0)['token0_out']) + (df1.fillna(0)['token1_in'] - df1.fillna(0)['token1_out']) *df1.Spotprice)

df['Apy']=(((1+df['fee']/df['Tvl'])**365)-1)*100
df['Apr']=(df['fee']/df['Tvl'])*365*100
```

```python
daily_first = df.groupby(df["block_timestamp"].dt.date).last().shift()
daily_last = df.groupby(df["block_timestamp"].dt.date).last()
daily_first.iloc[0]=df.iloc[0]
imp_loss=(abs(1-((2*(np.sqrt(daily_last['Spotprice']*daily_first['Spotprice'])))/(daily_last['Spotprice']+daily_first['Spotprice']))))*100
imp_loss=pd.DataFrame(imp_loss).reset_index()
imp_loss=imp_loss.rename(columns={'Spotprice': 'imp_loss','block_timestamp':'vaxt'})
imp_loss

```

```python
df['vaxt']=df['block_timestamp'].dt.date
df=pd.merge(df, imp_loss, on='vaxt', how='left')
df.drop('vaxt',inplace=True,axis=1)
df
```
