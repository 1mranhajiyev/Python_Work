## Subclass

- A subclass ( or derived class ) - extends the base class
- Use the parent class as a template, and add something else creating a new template
- Use **super()** to call parent's methods but you can also call them explicitly
- The `__init__` method which runs upon creation, problem is this overwrites the parents' `__init__` method, so we need to also call the parents' `__init__` method and pass it the name argument, else our cube will be nameless.
     - super().`__init__`(argument)
     - ParentName.`__init__`(self, argument)

```python
import random
```

```python
class Cube:
    """Parent Class"""
    def __init__(self, name):
        self.name = name
        self.color = 6
        
    def query(self):
        return('I am a cube, my name is: ' + self.name)
    
    def draft(self):
        return 'It"s a draft'
        
    def multiply_by_2(self, x):
        return x*2
        
    
class ColorCube(Cube):
    """Child Class"""
    def __init__(self, x):
        self.color = random.choice(['BLUE', 'RED', 'PURPLE'])
        Cube.__init__(self, x)
        
 
        
    def query(self):
        return super().query() + (' my color is: ' + self.color)
    
    def call_draft(self):
        return super().draft()
    
    def draft(self):
        print("it's not a draft") # bes burada print olsa?
        
    def print_answer(self, x):
        print(self.multiply_by_2(x)) # parent classin methodunu cagira bilerik?
        
    def print_draft_2times(self):
        return(self.draft()) # classin ozunun methodu bes?
        
    def print_name(self):
        print(self.color) # attributlari istifade ede bilerik?
    
colorCube1 = ColorCube('Bob')
# print(colorCube1.query())
# colorCube1.call_draft()
# print(colorCube1.draft())
colorCube1.print_answer(5)
# colorCube1.print_draft_2times()
# colorCube1.print_name()
```

## Multiple inheritance

- To subclass from 2 classes we add the class as an argument : class ColorCubeWithDimensions(Cube,Dimensions)
- If you are going to subclass from multiple classes, we can call the methods explicitly, but you can also use super():
      Cube.__init__(self,name)
      Dimensions.__init__(self)

```python
#     Parent class 1
class Cube:
    def __init__(self, name):
        self.name = "HAHA"
        
    def query(self):
        return('I am a cube, my name is: ' + self.name)
    
    
#     Parent class 2
class Dimensions:
    def __init__(self, name, width, height):
        self.name = 'x'
        self.width = 15
        self.height = 20
        
#     Child class      
class ColorCubeWithDimensions(Dimensions,Cube): # ardicilligin bir ferqi var?
    def __init__(self,name, width, height): # bu parametrler qeyd olunmasa ishleye biler?
        self.color = random.choice(['BLUE', 'RED', 'PURPLE'])
        Dimensions.__init__(self,name, width, height)
        Cube.__init__(self,name) # basqa nece yaza bilerik?
        
        
    def query(self):
        return  (' my color is: ' + self.color+"\n"+self.name ) + '\n' + ('My Dimensions are: WIDTH:' + str(self.width) + ' HEIGHT:' + str(self.height))

    
colorCube1 = ColorCubeWithDimensions('BOB',6,7)
print(colorCube1.query())
```

# Method overloading & overriding

 - Method Overloading is when more than one method of the same class shares the same method name having different signatures.
  - Note: Python does not support method overloading.
 - Method overriding is a feature by which the child class or the subclass can redefine methods present in the base class or parent class

```python
# Function to take multiple arguments
def add(datatype, *args):
    if datatype =='int':
        answer = 0

    if datatype =='str':
        answer =''

    for x in args:
       
        # This will do addition if the 
        # arguments are int. Or concatenation 
        # if the arguments are str
        answer = answer + x
        print(x, end = ' ')
        
    print('=', answer)
   
# Integer
add('int', 5, 6)
   
# String
add('str', 'Hi ', 'Geeks')
```

```python
class A:
    def fun1(self):
        print('feature_1 of class A')
         
    def fun2(self):
        print('feature_2 of class A')
     

class B(A): 
    # Modified function that is
    # already exist in class A
    def fun1(self):
        print('Modified feature_1 of class A by class B')   
         
    def fun3(self):
        print('feature_3 of class B')
         
 
# Create instance
obj = B()
     
# Call the override function
obj.fun1()
```
