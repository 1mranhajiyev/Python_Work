# Series

##### Homogeneous data
##### Size Immutable
##### Values of Data Mutable

```python
# Series is a one-dimensional array like structure with homogeneous data.
data = pd.Series([3, 8, 5, -3])
data
```

```python
# series are size immutable, which means when you add or delete value from series, it creates new series object
s1 = pd. Series([1, 2, 3])
s2 = pd. Series([4, 5, 6])
s1. append(s2)
print(s1)
```

```python
Series([5,8,9])
```

```python
data
```

```python
# index passed must be of the same length
data_2 = pd.Series([5, 2, -15, 33], index=['e', 'f', 'h','e'])
data_2
```

```python
# Create Series from dictionary
sdata = {'Baku': 35000, 'Ganja': 71000, 'Shamkir': 16000, 'Guba': 5000}

data_3 = pd.Series(sdata)
data_3
```

```python
print(data_3.values)
print(data_3.index)
```

```python
data_3.name = 'ehali'
data_3.index.name = 'sheher'
data_3
```

```python
# Index order is persisted and the missing element is filled with NaN
data4 = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data4,index=['b','c','d','a'])
print(s)
```

###### Accessing Data from Series

```python
# get a value
print(s[0])
```

```python
# get multiple values
print(s[:3])
print(s[-3:])
```

###### Retrieve Data Using Label (Index)

```python
# get a value
print(s['a'])
```

```python
# get multiple values
print(s[['a', 'b']])
```

```python
# getting non-existing key (index)
print(s['f'])
```
