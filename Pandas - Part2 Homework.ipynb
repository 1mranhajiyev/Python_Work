**Exercise 2:** Müxtəlif siniflərin cinsə və yaş aralıqlarına görə sağ qalma sayını göstərən pivot cədvəli qurun

Qeyd: Yaş kateqoriyaları 0 - 10; 10 - 30; 30 - 60; 60 - 80 və yaşı müəyyən edilməyənlər şəkilində olmalıdır.

```python

nm=[0,10,30,60,80]
res=titanic.pivot_table(values="Survived",index=["Sex",pd.cut(titanic["Age"],nm).cat.add_categories('Unknown').fillna('Unknown')],columns="Pclass",aggfunc='count')
res

# nm=[0,10,30,60,80]
# a=pd.cut(titanic["Age"],labels=nm,bins=5)
# a

```

**Exercise 3:** Titanikin sərnişinlərini əvvəlki çalışmadakı kimi yaş aralıqlarına görə qruplaşdırın və hər qrupda A kabinində olan sərnişinlərin sayını, maximum və minimum yaş fərqlərini əks etdirin

```python
a=titanic["Cabin"].str.startswith("A",na=False)
titanic[a]
indexing=pd.cut(titanic["Age"],nm).cat.add_categories('Unknown').fillna('Unknown')

res=titanic[a].pivot_table(index=indexing,aggfunc={'Pclass':'count', 'Age': lambda x: x.max() - x.min()}).rename({'Pclass':'Survived in class A',"Age":"Max Age-Min Age"}, axis=1)
res
# res
```

**Exercise 4:** Hər sinifdə sağ qalan qadın sərnişinlərin sayını tapın və onu yeni bir sütun olaraq Titanic cədvəlinə əlavə edin 

```python
maps = titanic[titanic['Sex'] == 'female'].pivot_table(index="Pclass",aggfunc={'Survived':'sum'})
mot = {}
maps.index
for i in maps.index:
    mot[i] = maps.Survived[i]
titanic['Survived females in class'] = titanic.Pclass.map(mot)
titanic
```
